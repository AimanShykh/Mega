```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame with multiple numerical columns


# Function to remove outliers using IQR
def remove_outliers_iqr(df, columns):
    df_clean = df.copy()
    for col in columns:
        if df_clean[col].dtype in ['int64', 'float64']:  # Ensure numeric columns
            Q1 = df_clean[col].quantile(0.25)
            Q3 = df_clean[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]

    return df_clean


# Boxplot before removing outliers
plt.figure(figsize=(25, 10))
sns.boxplot(data=dfle)
plt.title('Before Removing Outliers')
plt.show()


```

```python
# Remove outliers from all numerical columns
df_clean = remove_outliers_iqr(dfle, dfle.select_dtypes(include=['number']).columns)

```

```python
# Boxplot after removing outliers
plt.figure(figsize=(20, 10))
sns.boxplot(data=df_clean)
plt.title('After Removing Outliers')
plt.show()

```

```python
# Print the difference in data size
print(f'Original data size: {dfle.shape[0]} rows')
print(f'Cleaned data size: {df_clean.shape[0]} rows')
```

```python
X = df_clean.drop("normalized_used_price",axis='columns')
y = df_clean.normalized_used_price
```

```python
X=pd.get_dummies(X,columns=["device_brand","os"],drop_first=True)
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Import Models
from sklearn.linear_model import Ridge,Lasso

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
```

```python
def train_evaluate_model(model, model_name):
    pipeline = Pipeline([

        ("regressor", model)
    ])

    pipeline.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)

    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    print(f"{model_name} Performance:")
    print(f"  Train MSE: {train_mse:.4f} | Test MSE: {test_mse:.4f}")
    print(f"  Train R² Score: {train_r2:.4f} | Test R² Score: {test_r2:.4f}")
    print("-" * 50)

    return model, train_mse, test_mse, train_r2, test_r2
```

```python


linear_model = train_evaluate_model(LinearRegression(), "Linear Regression")
ridge_model = train_evaluate_model(Ridge(alpha=1.0), "Ridge Regression")
lasso_model = train_evaluate_model(Lasso(alpha=0.01), "Lasso Regression")
dt_model = train_evaluate_model(DecisionTreeRegressor(random_state=42), "Decision Tree")
rf_model = train_evaluate_model(RandomForestRegressor(n_estimators=100, random_state=42,max_depth=15), "Random Forest")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor #xgboost and polynomial regression
from sklearn.metrics import mean_squared_error

# Define hyperparameter grids
param_grids = {
    "Linear Regression": {},  # No hyperparameters for basic Linear Regression
    "Ridge Regression": {"alpha": [0.01, 0.1, 1, 10, 100]},
    "Lasso Regression": {"alpha": [0.0001, 0.001, 0.01, 0.1, 1]},
    "Decision Tree": {"max_depth": [5, 10, 15, 20], "min_samples_split": [2, 5, 10]},
    "Random Forest": {"n_estimators": [50, 100, 200], "max_depth": [10, 15, 20], "min_samples_split": [2, 5]}
}

# Define models
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42)
}

# Function to apply GridSearchCV
def train_with_gridsearch(model, param_grid, model_name, X_train, y_train, X_test, y_test):
    if param_grid:  # Only apply GridSearch if there are parameters
        grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
        grid_search.fit(X_train, y_train)
        best_model = grid_search.best_estimator_
        print(f"Best parameters for {model_name}: {grid_search.best_params_}")
    else:
        best_model = model.fit(X_train, y_train)  # No GridSearch for Linear Regression
    
    # Evaluate the model
    y_pred = best_model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f"{model_name} MSE: {mse:.4f}")
    return best_model

# Train and evaluate models
trained_models = {}
for name, model in models.items():
    trained_models[name] = train_with_gridsearch(model, param_grids[name], name, X_train, y_train, X_test, y_test)
```
